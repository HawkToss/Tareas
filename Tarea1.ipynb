```python
import random
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import csv

archivo = open('FallecidosEtario_T.csv') #Abrir archivo
reader = pd.read_csv(archivo)
reader = reader.values
Fechas=[]
Fechas2=[]
Muertes=[]
n=0


for row in reader:
    n=n+1
    x=row[0]
    y=int(row[1])+int(row[2])+int(row[3])+int(row[4])+int(row[5])+int(row[6])+int(row[7]) #Sumar todas las muertes de los distintos grupos etarios
    Fechas.append(x)
    if n/7==1: #Agrupar semanalmente
        Fechas2.append(x)
        Muertes.append(y)
        n=0

#Muertes por semana acumulado
fig, ax = plt.subplots()
ax.plot(Fechas2, Muertes, linestyle = "dashed")
plt.show()  



#Muertes por semana

Muertes2 = []
FD = 0
for i in range(len(Muertes)):
    FD=Muertes[i]-Muertes[i-1]
    Muertes2.append(FD)
    
Muertes2[0]=Muertes[0]
    
fig, ax = plt.subplots()
ax.plot(Fechas2, Muertes2, linestyle = "dashed")
plt.show()  


```

```python
import csv
Lista = []
n=0

#Abrir archivo 
with open('Defunciones.csv', newline='') as File:  
    reader = csv.reader(File, delimiter=';')
    for row in reader:
        Lista.append(row[1])
        Lista.append(row[12])

```

```python
#Contar la cantidad total de fechas para luego colocarlas en una nueva lista ordenadas en fecha ascendente
Lista2=[]
for i in Fechas:
    a=Lista.count(i)
    Lista2.append(a)
```

```python
%matplotlib inline
import pandas as pd
n=0
m=0
Lista3=[]
for i in Lista2:
    n=n+i
    m=m+1
    if m==7:
        Lista3.append(n)
        m=0
        n=0
        

#Grafico muertes totales por semana entre las fechas 2020-04-09 y 2021-05-01
fig, ax = plt.subplots()
ax.plot(Fechas2, Lista3, linestyle = "dotted")
plt.show()  


#Grafico muerte por covid a la semana y muertes totales entre las fechas 2020-04-09 y 2021-05-01
fig, ax = plt.subplots()
ax.plot(Fechas2, Muertes2,linestyle="dashed")
ax.plot(Fechas2, Lista3, linestyle = "dotted")
plt.show() 


#Se puede ver que el covid se comporta de distinta manera que la gripe, causando muchas mas muertes por semana

```
