```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
HDI=pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")
VAC=pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")
```

```python
SAN=VAC[VAC['Region']=='Metropolitana']
SAN
```

```python
col_list=list(SAN)
col_list.remove('Codigo region')
col_list.remove('Codigo comuna')
col_list.remove('Poblacion')
col_list.remove('Region')
col_list.remove('Comuna')



SAN=VAC[VAC["Region"].str.contains("Metropolitana")]
SAN=SAN.drop([140],axis=0)



SAN["Gente_Vacunada"] = SAN[col_list].sum(axis=1)
SAN
```

```python
HDI.rename(columns = {'comuna' : 'Codigo comuna'}, inplace = True)
MEZ=pd.merge(SAN,HDI,
how='inner', on='Codigo comuna')
```

```python
from pylab import *
x1 = MEZ['HDI']
y1 = MEZ['Gente_Vacunada']
x= np.array(x1)
y= np.array(y1)
m,b=polyfit(x,y,1)
plot(x,y,'yo', x, m*x+b, '--k')
show()
```
